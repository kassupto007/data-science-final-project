```python
#Heatmap of our data

fig, ax = plt.subplots(figsize=(12,6))
sns.heatmap(df.corr(), annot=True, cmap="Reds" )
```

__Inference:__ The above heatmap shows the correlation of numerical features in our data. The correlation coefficient is a measure ranging from -1 to 1 (with 1 being the maximum value and -1 being the smallest):<br>
* The Number closer to zero indicates a lower correlation (precise 0 implying no significant 
relation).
*	The Number around +1 to one indicates a more positive correlation.
*	The Number around -1 implies a more negative correlation.

### Boroughs with their collision numbers ###

```python
#Boroughs with their collision numbers
fig, ax = plt.subplots(figsize=(12,6))
sns.countplot(x=df['BOROUGH'], data=df, ax=ax)
plt.show()
```

## Inference:
- Brooklyn had the most number of collisions as compared to other boroughs while Staten Island had the least.

```python
#Adding a new column for years
df["CRASH DATE"] = pd.to_datetime(df["CRASH DATE"])
df["year"] = df["CRASH DATE"].dt.year
df["year"].unique()
```

```python
#Adding a month column for months
df["month"] = df["CRASH DATE"].dt.month
df["month"].unique()
```

### Checking for number of collisions by year ###

```python
#Checking for number of collisions by year
fig, ax = plt.subplots(figsize=(12,6))
sns.countplot(x=df['year'], data=df, ax=ax)
plt.show()
```

## Inference:
- As we can see from this graph, the year 2015 has the most number of collisions where the year 2021 has least.

### Checking for number of collisions by month ###

```python
#Checking for number of collisions by month
fig, ax = plt.subplots(figsize=(12,6))
sns.countplot(x=df['month'], data=df, ax=ax)
plt.show()
```

## Inference:
- As we can see here, the most number of collisions occured in July overall which is numbered as 7 in this graph.

### Boroughs with number of injuries

```python
#Boroughs with number of injuries
killed_by_borough = df.groupby("BOROUGH")[["NUMBER OF PERSONS INJURED", "NUMBER OF PEDESTRIANS INJURED",\
                                           "NUMBER OF CYCLIST INJURED", "NUMBER OF MOTORIST INJURED"]].sum()
print(killed_by_borough)
killed_by_borough.plot.bar(figsize=(12,6))
```

## Inference:
- As we can see from the above graph, most injuries occured in Brooklyn and Queens.

```python
persons_injured_by_year = df.groupby("year")["NUMBER OF PERSONS INJURED"].sum()
persons_injured_by_year
```

### Pie chart for persons killed by year ###

```python
#Pie chart for persons injured by year
fig, ax  = plt.subplots(figsize=(16, 8))
fig.suptitle('Persons injured by year', size = 14, color = "black")
ax = plt.pie(persons_injured_by_year, labels=persons_injured_by_year.index,
       autopct='%1.0f%%',
       pctdistance=0.9)
```

## Inference:
- From the above graph, we can see that the highest percentage of injuries occured in 2014 and 2013.

### Boroughs with number of deaths

```python
killed_by_borough = df.groupby("BOROUGH")[["NUMBER OF PERSONS KILLED", "NUMBER OF PEDESTRIANS KILLED",\
                                           "NUMBER OF CYCLIST KILLED", "NUMBER OF MOTORIST KILLED"]].sum()
print(killed_by_borough)
killed_by_borough.plot.bar(figsize=(12,6))
```

## Inference:
- As we can see from the above graph, most deaths occured in Brooklyn and Queens.

```python
persons_killed_by_year = df.groupby("year")["NUMBER OF PERSONS KILLED"].sum()
persons_killed_by_year
```

### Pie chart for persons killed by year ###

```python
#Pie chart for persons killed by year
fig, ax  = plt.subplots(figsize=(16, 8))
fig.suptitle('Persons killed by year', size = 14, color = "black")
ax = plt.pie(persons_killed_by_year, labels=persons_killed_by_year.index,
       autopct='%1.0f%%',
       pctdistance=0.9)
```

## Inference:
- From the above graph, we can see that the highest percentage of deaths occured in 2021, 2018 and 2016 respectively.
